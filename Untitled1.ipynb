```python
# 1. Λίστα (List): Ιδανική για σειρά δεδομένων (π.χ. σκορ ακρίβειας ενός AI μοντέλου)
model_scores = [85, 90, 92, 78, 88]

# 2. Λεξικό (Dictionary): Ιδανικό για ζεύγη "Κλειδί: Τιμή" (π.χ. ρυθμίσεις του μοντέλου)
model_config = {
    "algorithm": "Random Forest",
    "epochs": 100,
    "learning_rate": 0.01
}

# 3. Συνάρτηση (Function): Πακετάρουμε τη λογική μας για να την ξαναχρησιμοποιήσουμε
def evaluate_model(scores_list):
    good_scores = 0  # Μετρητής
    
    # 4. For Loop: Περνάμε μέσα από κάθε στοιχείο της λίστας
    for score in scores_list:
        
        # 5. If/Else (Έλεγχος Ροής)
        if score >= 85:
            good_scores += 1
        else:
            print(f"Προσοχή: Χαμηλό σκορ εντοπίστηκε ({score})")
            
    return good_scores

# 6. Καλούμε τη συνάρτηση δίνοντάς της τη λίστα μας
top_performances = evaluate_model(model_scores)

# Τυπώνουμε το τελικό αποτέλεσμα αντλώντας δεδομένα και από το Λεξικό
print(f"Το μοντέλο πέτυχε {top_performances} υψηλά σκορ χρησιμοποιώντας τον αλγόριθμο {model_config['algorithm']}.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Φορτώνουμε το dataset του Τιτανικού απευθείας από το seaborn
df = sns.load_dataset('titanic')

# 2. Το .head() μας δείχνει τις 5 πρώτες γραμμές του πίνακα
print("--- Οι 5 πρώτοι επιβάτες του Τιτανικού ---")
display(df.head())

# 3. Το .info() μας λέει πόσα δεδομένα έχουμε και αν λείπουν τιμές (missing data)
print("\n--- Πληροφορίες Dataset ---")
df.info()
```
