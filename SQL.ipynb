```python
import sqlite3
import pandas as pd

# 1. Σύνδεση με μια βάση δεδομένων (θα δημιουργηθεί ένα αρχείο bootcamp.db)
conn = sqlite3.connect('bootcamp.db')
cursor = conn.cursor()

# 2. SQL: Δημιουργία Πίνακα (Create Table)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY,
        name TEXT,
        track TEXT,
        score INTEGER
    )
''')

# 3. SQL: Εισαγωγή Δεδομένων (Insert)
students_data = [
    (1, 'Maria', 'AI Engineering', 95),
    (2, 'Giannis', 'Data Science', 88),
    (3, 'Eleni', 'AI Engineering', 92),
    (4, 'Kostas', 'Web Dev', 75)
]

cursor.executemany('INSERT OR REPLACE INTO students VALUES (?,?,?,?)', students_data)
conn.commit()

print("Η βάση δεδομένων δημιουργήθηκε και ενημερώθηκε!")
```

```python
# SQL Query: Θέλουμε μόνο τους μαθητές του AI Engineering με score > 90
query = "SELECT name, score FROM students WHERE track = 'AI Engineering' AND score > 90"

# Χρησιμοποιούμε το Pandas για να διαβάσουμε το αποτέλεσμα της SQL σε ωραίο πίνακα
df_sql = pd.read_sql_query(query, conn)

print("--- Αποτελέσματα SQL Query ---")
display(df_sql)

# Μην ξεχνάμε να κλείνουμε τη σύνδεση στο τέλος
conn.close()
```
