```python
# Ορισμός συνάρτησης (Function)
def check_ai_scores(scores_list):
    passed_students = {} # Dictionary για τα αποτελέσματα
    
    for i, score in enumerate(scores_list):
        # Control Flow (If/Else)
        if score >= 60:
            passed_students[f"Student_{i+1}"] = "Passed"
        else:
            passed_students[f"Student_{i+1}"] = "Failed"
            
    return passed_students

# Δεδομένα
test_scores = [45, 88, 92, 55, 76]

# Κλήση συνάρτησης
results = check_ai_scores(test_scores)
print(results)
```
